```python
def clean_data(raw_read):
    clean =raw_read.rename(columns={0:'Questions'})
    clean = clean.drop([0,1], axis=0)
    clean = clean.reset_index()
    clean = clean.drop('index', axis=1)
    clean['isAnswer'] = clean.Questions.str.find('ANSWER: ') & clean.Questions.str.find('ANSWERS: ') & clean.Questions.str.find('The correct answers ')
    clean['previous'] = clean.isAnswer.shift(1)
    clean['Options'] = clean[(clean.isAnswer==-1)&(clean.previous==-1)].Questions
    clean['Answers'] = clean[(clean.isAnswer==0)].Questions
    clean.Options = clean.Options.shift(-1)
    clean.Answers = clean.Answers.shift(-2)
    clean = clean.drop(['isAnswer','previous'], axis=1)
    clean = clean.fillna(0)
    drop_list = clean[(clean.Options==0)&(clean.Answers==0)].index.tolist()
    clean_final = clean.drop(drop_list, axis=0)
    return clean_final
```

```python
for i in range(2,5):
    raw_read = pd.read_csv(f'../input/additionalhuaweitests/HCIA-Cloud-Computing-Test-{i}-on-2020-1-jpg.txt',header=None, sep='\n')
    cleaned = clean_data(raw_read)
    cleaned.to_csv(f'computing_test{i}.csv')
```
